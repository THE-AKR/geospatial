<h1>TASK 4</h1>

```python
df=pd.read_csv("/content/parking_violations_2015.csv")
```

```python
df
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['lat','lon'],how='all',inplace=True)
```

```python
df.isna().sum()
```

```python
gdf=gpd.GeoDataFrame(df,geometry=gpd.points_from_xy(df.lon,df.lat),crs="EPSG:4326")
```

```python
gdf.head()
```

```python
gdf.to_file("parking.geojson",driver="GeoJSON")
```

```python
m.add_data(data=gdf,name="parking_points")
m
```
